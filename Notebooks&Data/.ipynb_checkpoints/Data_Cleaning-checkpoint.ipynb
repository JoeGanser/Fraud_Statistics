In this notebook I go through the process of formatting and cleaning the data into the structure it is used for analysis.

```python
import pandas as pd
df = pd.read_excel('united_states_offense_type_by_agency_2015.xlsx')
df.drop(['Unnamed: '+str(i) for i in [1,4,5,6,7,8]],axis=1,inplace=True)
labelsseries = df.iloc[3,:].dropna()
labels = dict(zip(labelsseries.index,labelsseries.values))
for key,value in labels.items():
    labels[key] = labels[key].replace('\n',' ')
labels['United States'] = 'State'
labels['Unnamed: 2'] = 'City'
labels['Unnamed: 3'] = 'Population'
df.rename(columns=labels,inplace=True)
df = df.loc[4:]
#df['total_crimes'] = df[[i for i in df.columns if i not in ['State','City','Population1']]].sum(axis=1)
not_crimes = ['State','City','Population']
crimes = [i for i in df.columns if i not in not_crimes]
df = df[not_crimes+crimes]
df.head()
```

'Homicide Offenses','Pros- titution Offenses' are both summary columns

```python
df.drop(['Homicide Offenses','Pros- titution Offenses'],axis=1,inplace=True)
```

'Fraud Offenses' is a summary column

```python
df[[i for i in df.columns if 'Fraud' in i]+['False Pretenses/ Swindle/ Confidence Game','Imper- sonation']].head()
df.drop('Fraud Offenses',axis=1,inplace=True)
```

'Larceny/ Theft Offenses' is a summary column.

```python
theft_crimes = ['Pocket- picking', 'Purse- snatching', 'Shop- lifting','Larceny/ Theft Offenses']+['Theft From Building', 'Theft From Coin Op- erated Machine or Device',
       'Theft From Motor Vehicle','Theft of  Motor  Vehicle Parts or Acces- sories', 'All Other Larceny']
df[theft_crimes]
df[[i for i in theft_crimes if i!='Larceny/ Theft Offenses']].sum(axis=1)
df.drop('Larceny/ Theft Offenses',axis=1,inplace=True)
```

'Drug/ Narcotic Offenses' is a summary column.

```python
df[['Drug/ Narcotic Offenses', 'Drug/ Narcotic Violations','Drug Equipment Violations']]
df.drop('Drug/ Narcotic Offenses',axis=1,inplace=True)
```

'Sex Offenses' is a summary column.

```python
df[['Sex Offenses','Rape', 'Sodomy', 'Sexual Assault With an Object', 'Fondling',
       'Sex Offenses, Non- forcible', 'Incest', 'Statutory Rape']]
df.drop('Sex Offenses',axis=1,inplace=True)
```

'Gambling Offenses' is a summary column

```python
df[['Gambling Offenses', 'Betting/ Wagering','Operating/ Promoting/ Assisting Gambling','Gambling Equipment Violations', 'Sports Tampering']].sum()
df.drop('Gambling Offenses',axis=1,inplace=True)
```

Now we notice that the states are listed, but inbetween the first listing of a state and the next listing are a bunch of NaNs. We want to be able to organize rows by state so the next line of code fixes this.

```python
df.dropna(subset=['Population'],inplace=True)
df = df.reset_index().drop('index',axis=1)
not_crimes = ['State','City','Population']
crimes = [i for i in df.columns if i not in not_crimes]
df['total_crimes'] = df[crimes].sum(axis=1)
df = df[not_crimes+['total_crimes']+crimes]
df.head()
```

```python
import numpy as np
from numpy import nan
states = list(df['State'])
for i in df.index:
    if (str(df.loc[i]['State']) == 'nan')&(str(states[i])=='nan'):
        k = list(df.loc[:i]['State'])
        last_state = [i for i in k if str(i)!='nan'][-1]
        states[i] = last_state
        
df['states']=pd.Series(states)
df.drop('State',axis=1,inplace=True)
df = df[['states']+[i for i in df.columns if 'state' not in i.lower()]]
df.head()
```

```python
df.to_csv('data_cleaned.csv')
df.head()
```
