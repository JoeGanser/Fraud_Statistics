```python
import pandas as pd

df = pd.read_csv('data_cleaned.csv',index_col=0)
numerical_cols = [i for i in df.columns if i not in ['states','Agency Name']]
df[numerical_cols] = df[numerical_cols].applymap(lambda x: float(x))
crimes = [i for i in df.columns if i not in ['states','Agency Name','total_crimes','Population1']]
df.head()
```

```python
df['Population1'].sum()
```

```python
#plot_crime = data[crimes].sum()


frauds = ['FraudOffenses','FalsePretenses/Swindle/ConfidenceGame','CreditCard/AutomatedTellerMachineFraud',
'Imper-sonation','WelfareFraud','WireFraud','Counter-feiting/Forgery']

other_crimes = [i for i in crimes if i not in frauds]
```

```python
df[other_crimes].sum()
```

```python
from plotly_api_key import plotly_api_key
import plotly
plotly.tools.set_credentials_file(username='JKGProfessional', api_key=plotly_api_key)

import plotly.plotly as py
import plotly.graph_objs as go



trace1 = go.Bar(
    y=frauds,
    x=df[frauds].sum(),
    name='Fraud crimes',
    orientation = 'h'
)
trace2 = go.Bar(
    y=other_crimes,
    x=df[other_crimes].sum(),
    name='Other crimes',
    orientation = 'h'
)

data = [trace2, trace1]
layout = go.Layout(
    barmode='group',
    autosize=True,
    width=1000,
    height=1000,
    showlegend=True,
    #xaxis=go.layout.XAxis(automargin=True,xaxis={"mirror" : "allticks", 'side': 'top'}),
    #yaxis=go.layout.YAxis(automargin=True),
    xaxis={"mirror" : "allticks", 'side': 'top'},
    margin=go.layout.Margin(l=300),
    title = go.layout.Title(
        text = 'Number of arrests for each crime across the entire US in 2015'))

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```
